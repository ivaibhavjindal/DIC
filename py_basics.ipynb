### To deal with importing and displaying images from Google Drive or PC to Google Colab. 

* Mount your Google Drive to Google Colab 

```
#  Mount your Google Drive to Google Colab 

from google.colab import drive
drive.mount('/content/gdrive')
```

*  Import an image from Google Drive

```
# Import an image from Google Drive 

from PIL import Image
import matplotlib.pyplot as plt

cat = Image.open('gdrive/My Drive/images/cats/1.jpg')
dog = Image.open('gdrive/My Drive/images/dogs/1.jpg')
```

* Display the image from Google Drive 

```
# Display the image from Google Drive  

plt.subplot(1, 2, 1)
plt.imshow(cat)
plt.subplot(1, 2, 2)
plt.imshow(dog)
```
